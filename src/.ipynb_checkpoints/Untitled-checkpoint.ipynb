```python
from scipy.spatial.distance import pdist, squareform, cdist
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib as plt
X = [(1,3,4), (4,5,6), (7,8,9)]
Y = [(1,3,4), (1,3,4), (1,3,4)]
X = np.array(X)
squareform(pdist(X))

G = nx.Graph()
G.add_edge((1,3,4),(4,5,6), weight=4.123)

G.add_edge((1,3,4),(7,8,9), weight=9.2736)
G.add_edge((4,5,6), (7,8,9), weight=5.196)
for n in G.neighbors((1,3,4)):
    print(G.edges[(1,3,4),n])

```
